```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

data_dir = './data/'
```

```python
f_app = os.path.join(data_dir,'application_train.csv')
df = pd.read_csv(f_app)

df.head(5)
```

#作業 <br />
請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP <br />
<br />
0 個小孩 <br />
有 1 - 2 個小孩 <br />
有 3 - 5 個小孩 <br />
有超過 5 個小孩 <br />
請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot <br />
請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 Z 轉換 後的分數 <br />

```python
cut_rule = np.array([0,1,2,3,5,float('inf')])
df['CNT_CHILDREN_GROUP'] = pd.cut(df.CNT_CHILDREN.values,cut_rule,include_lowest = True)
df.CNT_CHILDREN_GROUP.value_counts()
```

```python
grp = ['CNT_CHILDREN_GROUP','TARGET']
grouped_df = df.groupby(grp)['AMT_INCOME_TOTAL']
grouped_df.mean()
```

```python
#2-2
plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP','TARGET']
df.boxplot(column = plt_column, by = plt_by, showfliers = False, figsize=(15,15))
plt.suptitle(' ')
plt.show()
```

```python
#3

df['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x: (x-np.mean(x))/np.std(x) )
df[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head(5)
```
