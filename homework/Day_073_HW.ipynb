# 執行步驟
Step 1 : Initialize x =3. Then, find the gradient of the function, dy/dx = 2*(x+5).

Step 2 : Move in the direction of the negative of the gradient. We use a learning rate. Let us assume the learning rate → 0.01

Step 3 : Let’s perform 2 iterations of gradient descent

Step 4 : We can observe that the X value is slowly decreasing and should converge to -5 (the local minima). However, how many iterations should we perform?

```python
import numpy as np

cur_x = 3
lr = 0.01
precision = 0.000001
previos_step_size = 1
max_iter = 10000
iters = 0
df = lambda x : 2*(x+5)

iters_history = [iters]
x_history = [cur_x]
step_size = [previos_step_size]
```

```python
while previos_step_size > precision and iters < max_iter:
    prev_x = cur_x
    cur_x = cur_x - lr * df(cur_x)
    previos_step_size = abs(cur_x - prev_x)
    iters += 1
    print(f"Iteration : {iters} , x val : {cur_x}")
    step_size.append(prev_step_size)
    iters_history.append(iters)
    x_history.append(cur_x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(iters_history,x_history,'o-', ms = 3, lw = 1.5, color = 'b')
plt.xlabel('iters', fontsize = 16)
plt.ylabel('x vals', fontsize = 16)
plt.show()
```

```python
plt.plot(iters_history, step_size, 'o-', ms = 3, lw = 1.5, color = 'r')
plt.xlabel('iters', fontsize = 16)
plt.ylabel('step size',fontsize = 16)
plt.show()
```

```python
lrs = [0.1, 0.01, 0.001, 0.00001]
for lr in lrs:
    
    iters = 0
    previous_step_size = 1
    cur_x = 3
    
    iters_history = [iters]
    x_history = [cur_x]
    step_size = [previos_step_size]
    flag = True
    while previous_step_size > precision and iters < max_iter:
        prev_x = cur_x
        cur_x = cur_x - lr * df(cur_x)
        previous_step_size = abs(cur_x - prev_x)
        iters += 1
        iters_history.append(iters)
        x_history.append(cur_x)
        step_size.append(previous_step_size)

    plt.figure(figsize = (12,3))
    plt.subplot(1,2,1)
    plt.plot(iters_history, x_history, color = 'b')
    plt.xlabel('iters')
    plt.ylabel('x vals')
    plt.subplot(1,2,2)
    plt.plot(iters_history, step_size, color = 'r')
    plt.xlabel('iters')
    plt.ylabel('step size')
    plt.suptitle(f'learning rate = {lr}')
    plt.show()
```
