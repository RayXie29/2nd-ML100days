# Sigmoid
Sigmoid 數學函數表示方式 <br />
<br />
sigmoid = lambda x: 1 / (1 + np.exp(-x))<br />

```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid(x):
    return (1 / (1+np.exp(-x)))

def dsigmoid(x):
    return ( x * (1-x) )
x = np.linspace(-10, 10, 100)
plt.plot(x , sigmoid(x), 'b', label = 'linspace(-10,10,100)')
plt.grid()
plt.title('Sigmoid function')

#show sigmoid formula
plt.text(4,0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$' , fontsize = 15)

#resize the x and y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
plt.show()
```

# Softmax

```python

def softmax(x):
    return np.exp(x) / float(np.sum(exp(x)))

x = np.linspace(-5,5,100)
plt.plot(x, softmax(x), 'r')
plt.title('Softmax function')
plt.text(0.1,0.05, r'$\sigma(x)j=\frac{e^{zj}}{sum{e^{zj}}}$',fontsize = 15)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```

# Rectified Linear Unit- Relu
f(x)=max(0,x)<br />

```python

def ReLU(x):
    relu = lambda v : 0 if v<= 0 else v
    return np.array([relu(v) for v in x])
def dReLU(x):
    drelu = lambda v : 0 if v<= 0 else 1
    return np.array([drelu(v) for v in x])

x = np.linspace(-5,5,100)

plt.plot(x,ReLU(x),'r')
plt.text(-2,3, r'$\sigma(x)=np.max(0,x)$',fontsize = 15)
plt.title('ReLU function')
plt.grid()
plt.show()

plt.text(-5,0.55, r'$\sigma(x)=np.max(0,\frac{x}{abs(x)})$',fontsize = 12)
plt.plot(x,dReLU(x),'b')
plt.title('dReLU function')
plt.grid()
plt.show()
```
